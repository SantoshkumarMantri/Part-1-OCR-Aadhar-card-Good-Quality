```python
from tesserocr import PyTessBaseAPI
from PIL import Image
import re
```

```python
openimg= Image.open('D:/Downloads/Aadhaar-card.png')
grayimg= openimg.convert('L')
```

```python
openimg
```

```python
pixel= grayimg.load()
```

```python
for i in range(grayimg.size[0]):
    for j in range(grayimg.size[1]):
        if pixel[i,j]>150:
            pixel[i,j]=255
        else:
            pixel[i,j]=0
```

```python
grayimg.save('temp.png')
```

```python
with PyTessBaseAPI(path='D:/Tesseract-OCR/tessdata/.', lang='eng') as api:
    api.SetImageFile('temp.png')
    text= api.GetUTF8Text()
```

```python
textspl= text.split('\n')
```

```python
#extracting Aadhar Number
for i in range(len(textspl)):
    if re.search('[0-9]',textspl[i]):
        x= re.split('\s',textspl[i])
        UID= x[0]+' '+x[1]+' '+x[2]
```
